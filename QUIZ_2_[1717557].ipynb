# **STEP2: READING THE IMAGE lego.png**

The below code reads an image file and stores it as a grayscale image in the variable img. The image file is specified by the path '/content/lego.png'. The cv2.IMREAD_GRAYSCALE flag specifies that the image should be read in grayscale, rather than in color. The function cv2.imread is a part of the OpenCV library and is used for reading image files.

```python
img = cv2.imread('/content/lego.png', cv2.IMREAD_GRAYSCALE) #Load the image(read the image) 'lego.png' in grayscale using OpenCV library
```

The below code do the following:
1. Applies a threshold to the img image using the cv2.threshold function.
2. Creates a kernel with all elements set to 1 and of size (9, 10) using the np.ones function. The kernel is used for dilation.
3. Displays the mask image using plt.imshow and sets the colormap to "gray".

```python
_, mask = cv2.threshold(img, 150, 250, cv2.THRESH_BINARY_INV) #Transforms the image into a binary format
kernal = np.ones((9,10), np.uint8) #A matrix called 'kernel' is created and initialized in order to perform some operation
plt.imshow(mask, 'gray') #Displays the image using Matplotlib, with the binary values represented as different shades of gray."
```

The bellow code:
1. used the advantages of Morphological, becuse Morphological transformations are basic image processing techniques that operate on the shape of an image. They are typically applied to binary images and involve the use of a kernel, or structuring element, to define the nature of the transformation. The two main types of morphological transformations are erosion and dilation, and there are also various variations such as opening, closing, and gradient. These techniques can be used to modify the shape of objects in an image."  

2. Displays the image using Matplotlib, with the binary values represented as different shades of gray."

```python
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal) #closing small holes inside the foreground objects, or small black points on the object.
plt.imshow(closing, 'gray') #Displays the image using Matplotlib, with the binary values represented as different shades of gray."
```

## **Contour**
Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

In OpenCV, finding contours is like finding white object from black background. Object to be found should be white and background should be black.

```python
(cnt, heirarchy) = cv2.findContours(closing.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # There are three arguments in cv2.findContours() function, first one is source image, second is contour retrieval mode, third is contour approximation method. And it outputs the contours and hierarchy. 
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Converts image from BGR color space to RGB.
cv2.drawContours(rgb, cnt, -1, (150,255,0), 2) # Draws the contours around the objects
plt.imshow(rgb) #Displays the image using Matplotlib
     
```

```python
print('Number of object ', len(cnt)) # Counts the number of objects in the image
plt.show()
```
